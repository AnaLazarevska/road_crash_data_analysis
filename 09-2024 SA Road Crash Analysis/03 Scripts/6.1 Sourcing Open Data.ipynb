## 6. Data Wrangling and Consistency Checks for df_units

```python
# change column names

df_units = df_units.rename(columns={'Unit No': 'UNIT_NUMBER', 'No Of Cas': 'unit_number_of_casualties', 'Veh Reg State': 'unit_registration_state', 'Unit Type': 'unit_type', 'Veh Year': 'unit_registration_year', 'Direction Of Travel': 'direction_of_travel', 'Sex': 'unit_controller_sex', 'Age': 'unit_controller_age', 'Lic State': 'unit_controller_licence_state', 'Licence Class': 'unit_controller_licence_class', 'Licence Type': 'unit_controller_licence_type', 'Towing': 'towing', 'Unit Movement': 'unit_movement', 'Number Occupants': 'unit_number_of_occupants', 'Postcode': 'unit_controller_residential_postcode', 'Rollover': 'unit_rollover', 'Fire': 'unit_fire'})
```

```python
# unit_registration_state column

# change capitalisation of values

df_units['unit_registration_state'] = df_units['unit_registration_state'].replace({'UNKNOWN': 'Unknown', 'FEDERAL': 'Federal'})

# replace NaN values

df_units['unit_registration_state'] = df_units['unit_registration_state'].fillna('Not Applicable')

# if unit_type is Railway Vehicle the value of unit_registration_state should be Not Applicable

df_units.loc[(df_units['unit_registration_state'] == 'Unknown') & (df_units['unit_type'] == 'Railway Vehicle'), 'unit_registration_state'] = 'Not Applicable'
```

```python
# unit_type column

# change capitalisation

df_units['unit_type'] = df_units['unit_type'].replace({'RIGID TRUCK LGE GE 4.5T': 'Rigid Truck LGE GE 4.5T', 'OMNIBUS': 'Omnibus', 'SEMI TRAILER': 'Semi Trailer', 'BDOUBLE - ROAD TRAIN': 'BDouble - Road Train'})
```

```python
# unit_registration_year column

# if unit_registration_state is Not Applicable and unit_registration_year is NaN then unit_registration_year should be Not Applicable

df_units.loc[(df_units['unit_registration_state'] == 'Not Applicable') & (df_units['unit_registration_year'].isna()), 'unit_registration_year'] = 'Not Applicable'

# change XXXX to Unknown

df_units['unit_registration_year'] = df_units['unit_registration_year'].replace('XXXX', 'Unknown')
```

```python
# direction_of_travel column

# if unit_registration_state is Not Applicable and direction_of_travel is NaN then direction_of_travel should be Not Applicable

df_units.loc[(df_units['unit_registration_state'] == 'Not Applicable') & (df_units['direction_of_travel'].isna()), 'direction_of_travel'] = 'Not Applicable'
```

```python
# unit_controller_sex column

# if unit_registration_state is Not Applicable then unit_controller_sex should be Not Applicable

df_units.loc[(df_units['unit_registration_state'] == 'Not Applicable') & (df_units['unit_controller_sex'].isna()), 'unit_controller_sex'] = 'Not Applicable'

# if unit_number_of_occupants is 0.0 then unit_controller_sex should be Not Applicable

df_units.loc[(df_units['unit_number_of_occupants'] == 0.0) & (df_units['unit_controller_sex'].isna()), 'unit_controller_sex'] = 'Not Applicable'
```

```python
# unit_controller_age column

# if unit_controller_sex is Not Applicable then unit_controller_age should be Not Applicable

df_units.loc[(df_units['unit_controller_sex'] == 'Not Applicable') & ((df_units['unit_controller_age'] == 'X') | (df_units['unit_controller_age'] == 'XXX') | (df_units['unit_controller_age'].isna())), 'unit_controller_age'] = 'Not Applicable'

# if unit_controller_sex is Unknown then unit_controller_age should be Unknown

df_units.loc[(df_units['unit_controller_sex'] == 'Unknown') & ((df_units['unit_controller_age'] == 'X') | (df_units['unit_controller_age'] == 'XXX') | (df_units['unit_controller_age'].isna())), 'unit_controller_age'] = 'Unknown'

# change DD4 to 004 and XXX to Unknown

df_units['unit_controller_age'] = df_units['unit_controller_age'].replace({'DD4': '004', 'XXX': 'Unknown'})
```

```python
# unit_controller_licence_state column

# change capitalisation of values

df_units['unit_controller_licence_state'] = df_units['unit_controller_licence_state'].replace({'UNKNOWN': 'Unknown', 'FEDERAL': 'Federal'})

# if unit_controller_sex and unit_registration_state is Not Applicable then unit_controller_licence_state should be Not Applicable

df_units.loc[((df_units['unit_controller_sex'] == 'Not Applicable') | (df_units['unit_registration_state'] == 'Not Applicable')) & (df_units['unit_controller_licence_state'].isna()), 'unit_controller_licence_state'] = 'Not Applicable'

# if unit_controller_sex is Unknown then unit_controller_licence_state should be Unknown

df_units.loc[(df_units['unit_controller_sex'] == 'Unknown') & (df_units['unit_controller_licence_state'].isna()), 'unit_controller_licence_state'] = 'Unknown'

# change NaN values

df_units['unit_controller_licence_state'] = df_units['unit_controller_licence_state'].fillna('Unknown')
```

```python
# unit_controller_licence_class column

# if unit_controller_licence_state is Not Applicable then unit_controller_licence_class should be Not Applicable

df_units.loc[(df_units['unit_controller_licence_state'] == 'Not Applicable') & (df_units['unit_controller_licence_class'].isna()), 'unit_controller_licence_class'] = 'Not Applicable'

# if unit_controller_licence_state is Unknown then unit_controller_licence_class should be Unknown

df_units.loc[(df_units['unit_controller_licence_state'] == 'Unknown') & (df_units['unit_controller_licence_class'].isna()), 'unit_controller_licence_class'] = 'Unknown'

# change NaN values

df_units['unit_controller_licence_class'] = df_units['unit_controller_licence_class'].fillna('Unknown')
```

```python
# unit_controller_licence_type column

# if unit_controller_licence_state is Not Applicable then unit_controller_licence_type should be Not Applicable

df_units.loc[(df_units['unit_controller_licence_state'] == 'Not Applicable') & (df_units['unit_controller_licence_type'].isna()), 'unit_controller_licence_type'] = 'Not Applicable'

# if unit_controller_licence_state is Unknown then unit_controller_licence_type should be Unknown

df_units.loc[(df_units['unit_controller_licence_state'] == 'Unknown') & (df_units['unit_controller_licence_type'].isna()), 'unit_controller_licence_type'] = 'Unknown'

# change NaN values

df_units['unit_controller_licence_type'] = df_units['unit_controller_licence_type'].fillna('Unknown')
```

```python
# towing column

# if unit_registration_state is Not Applicable then towing should be Not Applicable

df_units.loc[(df_units['unit_registration_state'] == 'Not Applicable') & (df_units['towing'].isna()), 'towing'] = 'Not Applicable'

# if unit_registration_state is Unknown then towing should be Unknown

df_units.loc[(df_units['unit_registration_state'] == 'Unknown') & (df_units['towing'].isna()), 'towing'] = 'Unknown'

# change NaN values

df_units['towing'] = df_units['towing'].fillna('Not Applicable')
```

```python
# unit_movement column

# if unit_registration_state is Not Applicable then unit_movement should be Not Applicable

df_units.loc[(df_units['unit_registration_state'] == 'Not Applicable') & (df_units['unit_movement'].isna()), 'unit_movement'] = 'Not Applicable'

# change NaN values

df_units['unit_movement'] = df_units['unit_movement'].fillna('Unknown')
```

```python
# unit_number_of_occupants column

# if unit_registration_state is Not Applicable then unit_number_of_occupants should be Not Applicable

# df_units.loc[(df_units['unit_registration_state'] == 'Not Applicable') & (df_units['unit_number_of_occupants'].isna()), 'unit_number_of_occupants'] = 'Not Applicable'
```

```python
# unit_controller_residential_postcode column

# if unit_controller_sex is Unknown then unit_controller_residential_postcode should be Unknown

df_units.loc[((df_units['unit_controller_sex'] == 'Unknown') | (df_units['unit_controller_licence_state'] == 'Unknown')) & ((df_units['unit_controller_residential_postcode'].isna()) | (df_units['unit_controller_residential_postcode'] == 'XXX') | (df_units['unit_controller_residential_postcode'] == 'XXXX')), 'unit_controller_residential_postcode'] = 'Unknown'

# if unit_controller_sex is Not Applicable then unit_controller_residential_postcode should be Not Applicable

df_units.loc[((df_units['unit_controller_sex'] == 'Not Applicable') | (df_units['unit_controller_licence_state'] == 'Not Applicable')) & ((df_units['unit_controller_residential_postcode'].isna()) | (df_units['unit_controller_residential_postcode'] == 'XXX') | (df_units['unit_controller_residential_postcode'] == 'XXXX')), 'unit_controller_residential_postcode'] = 'Not Applicable'

# change remaining NaN, XXX, XXXX values

df_units['unit_controller_residential_postcode'] = df_units['unit_controller_residential_postcode'].fillna('Unknown')
df_units['unit_controller_residential_postcode'] = df_units['unit_controller_residential_postcode'].replace({'XXX': 'Unknown', 'XXXX': 'Unknown'})
```

```python
# unit_rollover column

# replace NaN

df_units['unit_rollover'] = df_units['unit_rollover'].fillna('N')
```

```python
# unit_fire column

# replace NaN

df_units['unit_fire'] = df_units['unit_fire'].fillna('N')
```

```python
# check for duplicate rows

df_units_dups = df_units[df_units.duplicated()]
df_units_dups
```

## 8. Data Wrangling and Consistency Checks for df_casualty

```python
# change column names

df_casualty = df_casualty.rename(columns={'UND_UNIT_NUMBER': 'UNIT_NUMBER', 'CASUALTY_NUMBER': 'casualty_number', 'Casualty Type': 'casualty_type', 'Sex': 'casualty_sex', 'AGE': 'casualty_age', 'Position In Veh': 'casualty_position_in_vehicle', 'Thrown Out': 'thrown_out', 'Injury Extent': 'injury_extent', 'Seat Belt': 'seat_belt', 'Helmet': 'helmet', 'Hospital': 'hospital'})
```

```python
# casualty_age column

# replace XXX and XX values

df_casualty['casualty_age'] = df_casualty['casualty_age'].replace({'XXX': 'Unknown', 'XX': 'Unknown'})
```

```python
# casualty_position_in_vehicle column

# replace NaN

df_casualty['casualty_position_in_vehicle'] = df_casualty['casualty_position_in_vehicle'].fillna('Not Applicable')
```

```python
# thrown_out column

# replace NaN

df_casualty['thrown_out'] = df_casualty['thrown_out'].fillna('Not Applicable')
```

```python
# seat_belt column

# replace NaN values based on certain conditions

df_casualty.loc[(df_casualty['seat_belt'].isna()) & ~((df_casualty['casualty_position_in_vehicle'] == 'Not Applicable') | (df_casualty['casualty_position_in_vehicle'] == 'Passenger of Motorcycle') | (df_casualty['casualty_type'] == 'Rider') | (df_casualty['casualty_type'] == 'Pedestrian')), 'seat_belt'] = 'Unknown'

# replace NaN

df_casualty['seat_belt'] = df_casualty['seat_belt'].fillna('Not Applicable')
```

```python
# helmet column

# replace NaN

df_casualty['helmet'] = df_casualty['helmet'].fillna('Not Applicable')
```

```python
# hospital column

# replace XXXXXX

df_casualty['hospital'] = df_casualty['hospital'].replace({'XXXXXX': 'Unknown'})

# change NaN values

df_casualty.loc[(df_casualty['hospital'].isna()) & ((df_casualty['injury_extent'] == 'Treated at Hospital') | (df_casualty['injury_extent'] == 'Admitted to Hospital')), 'hospital'] = 'Unknown'

# replace NaN

df_casualty['hospital'] = df_casualty['hospital'].fillna('Not Applicable')

# change capitalisation of values

df_casualty['hospital'] = df_casualty['hospital'].str.title()
```

```python
# check for duplicate rows

df_casualty_dups = df_casualty[df_casualty.duplicated()]
df_casualty_dups
```
